```python
class emply:
    def __init__(self,name,age , salary , gen):
        self.name= name
        self.age = age
        self.salary= salary
        self.gen=gen
        
    def show_emp_details(self):
        print(" name : " , self.name)
        print(" age : " , self.age)
        print(" salary : " , self.salary)
        print(" gender : " , self.gen)
        
```

Matplotlib :

> is a python library used for data visualization 


> you can create barplots, scatter-plots histograms and a lot more with matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# line plot 

x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
# adding title and lable 
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
# changing the line style 
# or changing the line aesthetics

plt.plot(x,y,color='g' , linestyle=':' , linewidth=2)
plt.show()
```

```python
# adding two lines in the same plot
x = np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='y' , linestyle=':' , linewidth=4)
plt.plot(x,y2,color='r' , linestyle='-' , linewidth=3)
plt.title(" Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g' , linestyle=':' , linewidth=2)

plt.subplot(1,2,2) # 1- one row 2- two col , 2 indetates the index 
plt.plot(x,y2,color='y' , linestyle=':' , linewidth=2)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='g' , linestyle=':' , linewidth=2)

plt.subplot(2,1,2) 
plt.plot(x,y2,color='y' , linestyle=':' , linewidth=2)

plt.show()
```
